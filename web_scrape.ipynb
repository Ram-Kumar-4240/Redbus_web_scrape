```python
from selenium import webdriver
from selenium.webdriver.common.by import By

def extract_div_content(url, div_locator):
    driver = webdriver.Chrome()  # Replace with your preferred webdriver
    driver.get(url)
    driver.implicitly_wait(10)
    div_text= []

    try:
        div_element = driver.find_element(By.XPATH, div_locator)  # Replace with appropriate locator
        div_text = div_element.text
    except Exception as e:
        print(f"Error extracting div content: {e}")
    finally:
        driver.quit()
    if div_text:
        return div_text
```

```python
url = "https://www.redbus.in"
div_locator = "//div[@class='sc-kjoXOD dtFSOe']/div[@class='sc-cHGsZl kbQVef']"  # Replace with the actual locator

bus_name = extract_div_content(url, div_locator)

bus_name = bus_name.split('\n')
bus_names = [bus_name[bus] for bus in range(0,len(bus_name),6)]
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

def extract_links_by_class(url, class_name):
    driver = webdriver.Chrome()  # Replace with your preferred webdriver
    driver.get(url)

    links = driver.find_elements(By.CSS_SELECTOR, f"a.{class_name}")
    route_links = []
    for link in links:
        href = link.get_attribute("href")
        if href:
            route_links.append(href)
    driver.quit()
    return route_links


rote = []
for name in bus_names:
    url_1 = f'{url}/online-booking/{name}'
    div_locator_1 = "route"
    rote.append(extract_links_by_class(url_1, div_locator_1))
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager

# Setup Chrome options
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run in headless mode (no GUI)

# Set up the WebDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

try:
    # Open the redBus website
    driver.get("https://www.redbus.in/bus-tickets/vijayawada-to-hyderabad")

    # Allow the page to load
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "p-left-10 icon icon-up")))

    # Locate and click the <i> tag using the class name or other locator
    i_tag = driver.find_element(By.CSS_SELECTOR, "div.butten i")
    i_tag.click()

    # Wait for the new <div> to appear (adjust the locator as necessary)
    new_div = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.new-data-div")))

    # Extract data from the new <div>
    data = new_div.text
    print("Data extracted from the new div:", data)

finally:
    # Clean up and close the browser
    driver.quit()
```
